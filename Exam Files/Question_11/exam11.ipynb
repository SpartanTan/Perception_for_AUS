```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('books.png')
b,g,r = cv2.split(img)
img = cv2.merge([r,g,b])
plt.imshow(img)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
blockSize = 2
apertureSize = 3
k = 0.04
t = 0
### Insert Harris Corner detector here ###
dst = cv2.cornerHarris(gray, blockSize, apertureSize, k)
### Draw the detected corners in the original image here ###
img[dst>0.01*dst.max()]=[255,0,0]
results = np.where(dst>0.01)
print(np.shape(results))
# Display image
plt. imshow(dst)
```
